```
# Load the dataset using Jupyter notebook
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/State_Drug_Utilization_Data_2010/State_Drug_Utilization_Data_2010.csv', sep=',')
```

```
# Show the .head() of the dataset and specify the number of rows to print, in this case 5.
df.head(5)
```

```
# Change the Quarter Begin Date column to the correct date format
df['Quarter Begin Date'] = pd.to_datetime(df['Quarter Begin Date'])
```

```
# Find the maximum date of your data and create a name (maxdate) for the max date.
maxdate=df['Quarter Begin Date'].max()
maxdate
```

```
# Subtract all the dates on the column from the maixmun date
for i in df['Quarter Begin Date']:
    print(maxdate-i)
```

```
# Convert the number of days into number of months
numberofmonths= df['Quarter Begin Date'].dt.strftime('%b')
numberofmonths
```

```
# Save the result on a new column (named Month) and add to the data set
df['Month'] = numberofmonths
df
```
