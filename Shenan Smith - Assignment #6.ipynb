```python
# Load the dataset using Jupyter notebook
# For this assignment, I'm using the stocks dataset
import pandas as pd
stocks = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/stock_data/stocks_yahoo.csv', sep=',')
```

```python
# Which rows of the df stocks_company contain a hyphen '-'?
stocks['company_name'].str.contains('-', regex=True)
```

```python
# In the column 'company_name', extract the stock symbol and add it to a new column named 'symbol'
stocks['symbol'] = stocks['company_name'].str.extract('([A-Z]\w{0,})', expand=True)
stocks['symbol']
```

```python
# In the column 'company_name', extract only the company name without the symbol and add it to a new column named 'company_name_cleaned'
stocks['company_name_cleaned'] = stocks['company_name'].str.split('-').str[1]
stocks['company_name_cleaned']
```

```python
# Remove the rows that have missing values on the price_at_close and previous_close columns
stocks_cleaned = stocks.dropna(subset=['price_at_close', 'previous_close'])
stocks_cleaned.isnull().sum()
```

```python
# Separate the price_at_close and previous_close columns
price = stocks_cleaned[['price_at_close', 'previous_close']]
price
```
