```
# Load the dataset using Jupyter notebook
# Info about the data can be found here: https://www.medicaid.gov/medicaid/prescription-drugs/state-drug-utilization-data/state-drug-utilization-data-faq/index.html?search_api_fulltext=91956
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/State_Drug_Utilization_Data_2010/State_Drug_Utilization_Data_2010.csv', sep=',')
```

```
# Show the .head() of the dataset and specify the number of rows to print, in this case 5.
df.head(5)
```

```
# Load matplotlib and seaborn packages
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Chart #1
# This count plot, or bar plot, shows how many times each State had a drug utilization for a covered outpatient drug

# Change the dimensions of the graph to show all 50 states
count_dims = (18,12)
count, ax = plt.subplots(figsize=count_dims)

ax = sns.countplot('State', ax=ax,data=df)
# Add the title to the graph
ax.set_title('Count of State')
# Specify which variable to use on the x-axis, in this case State
ax.set_xlabel('State')
# Specify how to count the totals on the y-axis, in this case by frequency
ax.set_ylabel('Frequency')
# Show the graph
plt.show()
# A count plot, or bar plot, is best used when you want to count discrete variables.  
```

```
# Chart #2
# This bar plot shows the total amount of units reimbursed by state
# Change the dimensions of the graph to fit all 50 states, along with the XX variable, which represents the National Totals 
bar_dims = (18,12)
bar, ax = plt.subplots(figsize=bar_dims)
# Make the graph
sns.barplot(x=df['State'],y=df['Units Reimbursed'],ax=ax,data=df)
# Set the title as Total Units Reimbursed per State
ax.set_title('Total Units Reimbursed per State')
plt.show()
```

```
# Scatterplot of Total Amount Reimbursed compared to Number of Prescriptions
scatter, ax = plt.subplots()
ax = sns.regplot (x='Total Amount Reimbursed', y='Number of Prescriptions', data=df)
ax.set_title('Total Amount Reimbursed versus Number of Prescriptions')
ax.set_xlabel('Total Amount Reimbursed')
ax.set_ylabel('Number of Prescriptions')
plt.show()
```
